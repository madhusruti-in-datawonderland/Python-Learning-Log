```python
a = (input("Enter a number"))              #here the numbers are treated as 'strings' in the input functions. So the result is getting concatinated.
b = (input("Enter another number"))
print("Addition of the two numbers is", a+b)
```

```python
a = int(input("Enter a number"))      #in order to get the addition, we have to define the datatypes of the numbers in the input functions i.e. 'int'.
b = int(input("Enter another number"))
print("Addition of the two numbers is", a+b)
print("Substraction of the gtwo numbers is", a-b)
print("Multiplication of the two numbers is", a*b)
print("Division of the two numbers is", a/b)
```

## Typecasting --> Changing the datatype

## Implicit Typecasting (Changing the datatype, not by force)

```python
mm = 98
type(mm)
```

```python
mm = mm + 0.7
type(mm)
```

## Explicit Typecasting (Changing the datatype, by force)

```python
nn = "53"
type(nn)
```

```python
nn = int(nn)
type(nn)
```

## F-strings (formatted string literals)

```python
name = input("enter the name")
print(f"Hello! We are {name}.")
```

```python
first_name = input("enter the first name")
last_name = input("enter the last name")
print(f"Hi! My name is {first_name} {last_name}.")
```

## Modulus Operator (Remainder --> %) 

```python
# Whether a number is even or odd.
X = int(input("enter the number"))
if X%2 == 0:                          # Equality Operator
    print("Even")
else:
    print("Odd")
```
